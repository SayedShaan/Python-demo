```python
import pandas as pd
df=pd.read_csv("company-sales.csv")
df.head(5)
```

```python
import matplotlib.pyplot as plt
plt.plot(df.month_number,df.total_profit ,linestyle="dashed",color="red",
        marker="o",markerfacecolor ="black", linewidth="3",label = "Profit data of last year")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Company Sales data of last Year")
plt.legend(loc="lower right")
```

```python
import pandas as pd
df=pd.read_csv("company-sales.csv")
df.head(5)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(df.month_number,df.facecream,marker="o",linewidth="3",label="Face Cream Sales Data")
plt.plot(df.month_number,df.facewash,marker="o",linewidth="3",label="Face Wash Sales Data")
plt.plot(df.month_number,df.toothpaste,marker="o",linewidth="3",label="Toothpaste Sales Data")
plt.plot(df.month_number,df.bathingsoap,marker="o",linewidth="3",label="Bathingsoap Sales Data")
plt.plot(df.month_number,df.shampoo,marker="o",linewidth="3",label="Shampoo Sales Data")
plt.plot(df.month_number,df.moisturizer,marker="o",linewidth="3",label="Moisturizer Sales Data")
plt.legend(loc="upper left")
```

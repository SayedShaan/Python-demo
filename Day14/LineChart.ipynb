```python
import matplotlib.pyplot as plt

# Data for week1 and week2
week1 = ["mon", "tue", "wed", "thu", "fri", "sat", "sunday"]
temp1 = [27, 26, 28, 27, 29, 29, 30]

week2 = ["mon", "tue", "wed", "thu", "fri", "sat", "sunday"]
temp2 = [25, 26, 27, 27, 28, 27, 28]

# Plotting the data
plt.plot(week1, temp1, color="red", label="Week1")
plt.plot(week2, temp2, color="blue", label="Week2")

# Adding title and labels
plt.title("Week1 vs Week2")
plt.xlabel("Weekdays")
plt.ylabel("Temperature")

# Displaying the legend
plt.legend()

# Showing the plot
plt.show()
```

```python
import pandas as pd 
```

```python
df = pd.read_csv("student_details.csv")
df
```

```python
df["full_name"]=df["first_name"]+" "+df["last_name"]
df
```

```python
df["Total_marks"]=df["subject1"]+df["subject2"]+df["subject3"]
df
```

```python
df["Percentage"]=df["Total_marks"]/300*100
df
```

```python
df["Status"]=df["Percentage"].apply(lambda x:"Pass" if x>=50 else "Fail")
df
```

```python
def getstatus(per):
    if(per>=50):
        return "pass"
    else:
        return "fail"
df["Status"]=df["Percentage"].apply(getstatus)
df
```

```python
def getGrade(per):
    if(per>=75):
        return "Grade A"
    elif(per>=65):
        return "Grade B"
    elif(per>=55):
        return "Grade C"
    elif(per>=45):
        return "Grade D"
    else:
        return "Fail"

df["Grade"]=df["Percentage"].apply(getGrade)
df
```
